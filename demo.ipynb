```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pixels = []
with open('img1.csv', 'r') as f:
    for l in f.readlines():
        pixels.append(int(l.rstrip()))
print(len(pixels))        
```

```python
import numpy as np
img = np.reshape(pixels, (28,28))
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```
